```python
# looking at your data
df.describe()
```

```python
# making changes to your data # adding a column
df['Total Strength'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def']
df.head(5)
```

```python
# dropping a column
df = df.drop(columns = 'Total Strength')
df.head(5)
```

```python
# adding rows and making a column
df['Max Strength'] = df.iloc[:, 4:10].sum(axis = 1)
df.head(5)
```

```python
# reordering the columns
# create a list of the column names
cols_list = list(df.columns)
new_df = df[cols_list[:4] + cols_list[5:8]]
new_df.head(5)
```

```python
print(df.columns.get_loc('Type 1'))
```

```python
# getting location and creating dataframe
new_df = df[cols_list[df.columns.get_loc('#'):df.columns.get_loc('Generation')]]
new_df.head(5)
```

```python
# getting rid of the index
new_df.to_csv('modified_data.csv', index = False)
```
